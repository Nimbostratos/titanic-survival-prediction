```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_pd = pd.read_csv('//kaggle/input/titanictrain/train.csv')
test_pd = pd.read_csv('/kaggle/input/titanictest/test.csv')

display(train_pd)
```

```python
survived = train_pd['Survived']
train_pd = train_pd.drop(['Survived'], axis=1)
display(train_pd)
```

```python
train_idx = train_pd['PassengerId']
test_idx = test_pd['PassengerId']

combined_pd = pd.concat([train_pd, test_pd]).reset_index(drop=True)

display(combined_pd)
```

```python
# Exploratory Data Analysis - Survival Patterns
import seaborn as sns

# Recreate full training data with survival for EDA
train_with_survival = train_pd.copy()
train_with_survival['Survived'] = survived

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('Exploratory Analysis - Survival Rates', fontsize=16, fontweight='bold')

# 1. Survival rate by Sex
sex_survival = train_with_survival.groupby('Sex')['Survived'].agg(['sum', 'count', 'mean'])
ax1 = axes[0, 0]
sex_survival['mean'].plot(kind='bar', ax=ax1, color=['#e74c3c', '#3498db'])
ax1.set_title('Survival Rate by Sex', fontsize=12, fontweight='bold')
ax1.set_ylabel('Survival Rate')
ax1.set_xlabel('Sex')
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=0)
for i, v in enumerate(sex_survival['mean']):
    ax1.text(i, v + 0.02, f'{v:.1%}', ha='center', fontweight='bold')

# 2. Survival rate by Pclass
pclass_survival = train_with_survival.groupby('Pclass')['Survived'].agg(['sum', 'count', 'mean'])
ax2 = axes[0, 1]
pclass_survival['mean'].plot(kind='bar', ax=ax2, color=['#2ecc71', '#f39c12', '#e74c3c'])
ax2.set_title('Survival Rate by Class', fontsize=12, fontweight='bold')
ax2.set_ylabel('Survival Rate')
ax2.set_xlabel('Class')
ax2.set_xticklabels(['1st', '2nd', '3rd'], rotation=0)
for i, v in enumerate(pclass_survival['mean']):
    ax2.text(i, v + 0.02, f'{v:.1%}', ha='center', fontweight='bold')

# 3. Survival rate by Embarked
embarked_survival = train_with_survival.groupby('Embarked')['Survived'].agg(['sum', 'count', 'mean']).dropna()
ax3 = axes[0, 2]
embarked_survival['mean'].plot(kind='bar', ax=ax3, color=['#9b59b6', '#1abc9c', '#34495e'])
ax3.set_title('Survival Rate by Port', fontsize=12, fontweight='bold')
ax3.set_ylabel('Survival Rate')
ax3.set_xlabel('Port of Embarkation')
ax3.set_xticklabels(ax3.get_xticklabels(), rotation=0)
for i, v in enumerate(embarked_survival['mean']):
    ax3.text(i, v + 0.02, f'{v:.1%}', ha='center', fontweight='bold')

# 4. Age distribution by survival
ax4 = axes[1, 0]
train_with_survival[train_with_survival['Survived']==1]['Age'].dropna().hist(
    bins=20, ax=ax4, alpha=0.6, label='Survived', color='#2ecc71', edgecolor='black'
)
train_with_survival[train_with_survival['Survived']==0]['Age'].dropna().hist(
    bins=20, ax=ax4, alpha=0.6, label='Did not survive', color='#e74c3c', edgecolor='black'
)
ax4.set_title('Age Distribution by Survival', fontsize=12, fontweight='bold')
ax4.set_xlabel('Age')
ax4.set_ylabel('Frequency')
ax4.legend()

# 5. Survival count
ax5 = axes[1, 1]
survival_counts = train_with_survival['Survived'].value_counts()
ax5.bar(['Did not survive', 'Survived'], survival_counts.values, color=['#e74c3c', '#2ecc71'])
ax5.set_title('Total Survival Count', fontsize=12, fontweight='bold')
ax5.set_ylabel('Number of Passengers')
for i, v in enumerate(survival_counts.values):
    ax5.text(i, v + 10, str(v), ha='center', fontweight='bold')

# 6. Fare distribution by survival
ax6 = axes[1, 2]
survived_fare = train_with_survival[train_with_survival['Survived']==1]['Fare'].dropna()
died_fare = train_with_survival[train_with_survival['Survived']==0]['Fare'].dropna()
ax6.hist(died_fare, bins=30, alpha=0.6, label='Did not survive', color='#e74c3c', range=(0, 300))
ax6.hist(survived_fare, bins=30, alpha=0.6, label='Survived', color='#2ecc71', range=(0, 300))
ax6.set_title('Survival Rate Distribution', fontsize=12, fontweight='bold')
ax6.set_xlabel('Rate ($)')
ax6.set_ylabel('Frequency')
ax6.legend()

plt.tight_layout()
plt.show()

# Summary statistics
print("\n" + "="*60)
print("SUMMARY OF SURVIVAL RATES")
print("="*60)
print(f"\nOverall survival rate: {survived.mean():.1%}")
print(f"\nBy Sex:\n{sex_survival[['count', 'mean']]}")
print(f"\nBy Class:\n{pclass_survival[['count', 'mean']]}")
print(f"\nBy Port:\n{embarked_survival[['count', 'mean']]}")
print("="*60)
```

```python
# Feature Correlation Analysis
import seaborn as sns

# Recreate training data with survival for correlation
train_corr = combined_pd.loc[train_idx].copy()
train_corr['Survived'] = survived.values

# Encode categorical variables for correlation
train_corr_encoded = train_corr.copy()

# Encode categorical columns
categorical_cols = ['Sex', 'Embarked', 'Cabin', 'Title', 'Age_Bin', 'Fare_Bin']
for col in categorical_cols:
    if col in train_corr_encoded.columns:
        train_corr_encoded[col] = train_corr_encoded[col].astype('category').cat.codes

# Calculate correlation matrix
correlation_matrix = train_corr_encoded.corr()

# Plot correlation heatmap
plt.figure(figsize=(14, 12))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', 
            center=0, square=True, linewidths=0.5,
            cbar_kws={"shrink": 0.8})
plt.title('Feature Correlation Matrix', fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()

# Correlation with target variable (Survived)
survival_corr = correlation_matrix['Survived'].sort_values(ascending=False)

print("\n" + "="*60)
print("CORRELATION WITH SURVIVAL")
print("="*60)
print("\nCharacteristics most correlated with survival:")
print(survival_corr.head(10))
print("\nCharacteristics less correlated with survival:")
print(survival_corr.tail(10))
print("="*60)

# Visualize survival correlation
fig, ax = plt.subplots(figsize=(10, 8))
survival_corr_filtered = survival_corr[survival_corr.index != 'Survived']
colors = ['#2ecc71' if x > 0 else '#e74c3c' for x in survival_corr_filtered.values]
survival_corr_filtered.plot(kind='barh', ax=ax, color=colors)
ax.set_title('Correlation of Traits with Survival', fontsize=14, fontweight='bold')
ax.set_xlabel('Correlation Coefficient')
ax.axvline(x=0, color='black', linestyle='-', linewidth=0.8)
plt.tight_layout()
plt.show()

# Check for multicollinearity (features highly correlated with each other)
print("\n" + "="*60)
print("MULTICOLINEARITY (High correlations between characteristics)")
print("="*60)
high_corr_pairs = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        if abs(correlation_matrix.iloc[i, j]) > 0.7:
            high_corr_pairs.append((
                correlation_matrix.columns[i],
                correlation_matrix.columns[j],
                correlation_matrix.iloc[i, j]
            ))

if high_corr_pairs:
    for feat1, feat2, corr_val in high_corr_pairs:
        print(f"{feat1} <-> {feat2}: {corr_val:.3f}")
else:
    print("No correlated feature pairs were found > 0.7")
print("="*60)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import train_test_split, RandomizedSearchCV, GridSearchCV
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import MinMaxScaler
```

```python
combined_pd.set_index('PassengerId', inplace=True)
display(combined_pd)
```

```python
one_hot = pd.get_dummies(combined_pd.copy(), columns=['Sex', 'Embarked','Title'], dtype=int)
display(one_hot)
```

```python
# Make sure one_hot really is one-hot encoded and numeric
x = pd.get_dummies(one_hot.loc[train_idx], drop_first=True).values
y = survived.values

scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x)

x_train, x_test, y_train, y_test = train_test_split(
    x_scaled, y, test_size=0.2, random_state=0
)

print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
params = {
    'n_estimators': [50, 100, 200, 300, 400, 500, 800, 1000],
    'learning_rate': [1.0, 0.1, 0.01, 0.001]
}

rand_search = RandomizedSearchCV(
    AdaBoostClassifier(random_state=0),
    params,
    scoring='accuracy',
    random_state=0,
    cv=5
)

rand_search.fit(x_train, y_train)
rand_params = rand_search.best_params_

print("Random Search best params:", rand_params, '\n')
print("Train Acc:", rand_search.best_score_)
preds = rand_search.predict(x_test)
print("Test Acc:", accuracy_score(y_test, preds))
```

```python
import numpy as np

n_estimators = np.linspace(rand_params['n_estimators'] - 50,
                           rand_params['n_estimators'] + 50, 3).astype(int)
learning_rate = np.linspace(rand_params['learning_rate'] * 0.9,
                            rand_params['learning_rate'] * 1.1, 3)

params = {
    'n_estimators': n_estimators,
    'learning_rate': learning_rate
}

grid_search = GridSearchCV(
    AdaBoostClassifier(random_state=0),
    params,
    scoring='accuracy',
    cv=5
)

grid_search.fit(x_train, y_train)
grid_params = grid_search.best_params_

print("Grid Search best params:", grid_params, '\n')
print("Train Acc:", grid_search.best_score_)
preds = grid_search.predict(x_test)
print("Test Acc:", accuracy_score(y_test, preds))
```

```python
columns = one_hot.columns
best_model = grid_search.best_estimator_

train_acc = grid_search.best_score_
preds = best_model.predict(x_test)
test_acc = accuracy_score(y_test, preds)

print("Best hyperparameters:\n", grid_search.best_params_)
print("Train Acc:", train_acc)
print("Test Acc:", test_acc)
```

```python
# Enhanced Model Evaluation
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc, precision_recall_curve
from sklearn.model_selection import cross_val_score

# Get predictions and probabilities
y_pred = best_model.predict(x_test)
y_pred_proba = best_model.predict_proba(x_test)[:, 1]

# 1. Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Plot confusion matrix
ax1 = axes[0]
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax1, cbar=False,
            xticklabels=['Did not survive', 'Survived'],
            yticklabels=['Did not survive', 'Survived'])
ax1.set_title('Confusion Matrix', fontsize=14, fontweight='bold')
ax1.set_ylabel('Real Value', fontsize=12)
ax1.set_xlabel('Prediction', fontsize=12)

# Add percentages
total = cm.sum()
for i in range(2):
    for j in range(2):
        percentage = cm[i, j] / total * 100
        ax1.text(j+0.5, i+0.7, f'({percentage:.1f}%)', 
                ha='center', va='center', fontsize=10, color='gray')

# 2. ROC Curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
roc_auc = auc(fpr, tpr)

ax2 = axes[1]
ax2.plot(fpr, tpr, color='#3498db', lw=2, label=f'ROC curve (AUC = {roc_auc:.3f})')
ax2.plot([0, 1], [0, 1], color='gray', lw=1, linestyle='--', label='Random')
ax2.set_xlim([0.0, 1.0])
ax2.set_ylim([0.0, 1.05])
ax2.set_xlabel('False Positive Rate', fontsize=12)
ax2.set_ylabel('True Positive Rate', fontsize=12)
ax2.set_title('ROC curve', fontsize=14, fontweight='bold')
ax2.legend(loc="lower right")
ax2.grid(alpha=0.3)

# 3. Precision-Recall Curve
precision, recall, _ = precision_recall_curve(y_test, y_pred_proba)
pr_auc = auc(recall, precision)

ax3 = axes[2]
ax3.plot(recall, precision, color='#e74c3c', lw=2, label=f'PR curve (AUC = {pr_auc:.3f})')
ax3.set_xlim([0.0, 1.0])
ax3.set_ylim([0.0, 1.05])
ax3.set_xlabel('Recall', fontsize=12)
ax3.set_ylabel('Precision', fontsize=12)
ax3.set_title('Precision-Recall Curve', fontsize=14, fontweight='bold')
ax3.legend(loc="lower left")
ax3.grid(alpha=0.3)

plt.tight_layout()
plt.show()

# Classification Report
print("\n" + "="*70)
print("CLASSIFICATION REPORT")
print("="*70)
print(classification_report(y_test, y_pred, 
                          target_names=['Did not survive', 'Survived'],
                          digits=3))

# Cross-validation scores on full training set
print("="*70)
print("CROSS VALIDATION (5-fold)")
print("="*70)
cv_scores = cross_val_score(best_model, x_scaled, y, cv=5, scoring='accuracy')
print(f"Scores per fold: {cv_scores}")
print(f"Media: {cv_scores.mean():.4f}")
print(f"Standard deviation: {cv_scores.std():.4f}")
print(f"Confidence interval 95%: [{cv_scores.mean() - 1.96*cv_scores.std():.4f}, "
      f"{cv_scores.mean() + 1.96*cv_scores.std():.4f}]")
print("="*70)

# Visualize CV scores
plt.figure(figsize=(10, 6))
plt.boxplot([cv_scores], labels=['AdaBoost'])
plt.scatter([1]*len(cv_scores), cv_scores, alpha=0.6, s=100, c='#3498db')
plt.axhline(y=cv_scores.mean(), color='#e74c3c', linestyle='--', 
            label=f'Media: {cv_scores.mean():.4f}')
plt.ylabel('Accuracy', fontsize=12)
plt.title('Distribution of Cross-Validation Scores', fontsize=14, fontweight='bold')
plt.ylim([cv_scores.min() - 0.02, cv_scores.max() + 0.02])
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# Error Analysis
print("\n" + "="*70)
print("ERROR ANALYSIS")
print("="*70)
false_positives = np.where((y_test == 0) & (y_pred == 1))[0]
false_negatives = np.where((y_test == 1) & (y_pred == 0))[0]

print(f"\nFalse Positives (predicted survived, but died): {len(false_positives)}")
print(f"False Negatives (predicted he died, but survived): {len(false_negatives)}")
print(f"\nError rate: {(len(false_positives) + len(false_negatives)) / len(y_test):.1%}")
print("="*70)
```

```python
# Model Comparison
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import cross_val_score
import time

print("="*70)
print("MODEL COMPARISON")
print("="*70)
print("\nTraining multiple models...\n")

models = {
    'AdaBoost (optimized)': best_model,
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=0),
    'Gradient Boosting': GradientBoostingClassifier(n_estimators=100, random_state=0),
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=0),
    'SVM': SVC(kernel='rbf', probability=True, random_state=0),
    'K-Nearest Neighbors': KNeighborsClassifier(n_neighbors=5),
    'Naive Bayes': GaussianNB()
}

# Store results
results = []

# Train and evaluate each model
for name, model in models.items():
    start_time = time.time()
    
    # Cross-validation scores
    cv_scores = cross_val_score(model, x_scaled, y, cv=5, scoring='accuracy')
    
    # Train on full training set and test
    if name != 'AdaBoost (optimized)':  # Already trained
        model.fit(x_train, y_train)
    
    train_score = model.score(x_train, y_train)
    test_score = model.score(x_test, y_test)
    
    elapsed_time = time.time() - start_time
    
    results.append({
        'Model': name,
        'CV Mean': cv_scores.mean(),
        'CV Std': cv_scores.std(),
        'Train Acc': train_score,
        'Test Acc': test_score,
        'Time (s)': elapsed_time
    })
    
    print(f"âœ“ {name:25s} | CV: {cv_scores.mean():.4f}Â±{cv_scores.std():.4f} | "
          f"Test: {test_score:.4f} | Time: {elapsed_time:.2f}s")

# Create results dataframe
results_df = pd.DataFrame(results)
results_df = results_df.sort_values('CV Mean', ascending=False)

print("\n" + "="*70)
print("SUMMARY OF RESULTS")
print("="*70)
display(results_df.style.highlight_max(subset=['CV Mean', 'Test Acc'], color='lightgreen')
                        .highlight_min(subset=['CV Std', 'Time (s)'], color='lightgreen')
                        .format({'CV Mean': '{:.4f}', 'CV Std': '{:.4f}', 
                                'Train Acc': '{:.4f}', 'Test Acc': '{:.4f}', 
                                'Time (s)': '{:.2f}'}))

# Visualize comparison
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Plot 1: Cross-validation scores comparison
ax1 = axes[0]
x_pos = np.arange(len(results_df))
ax1.barh(x_pos, results_df['CV Mean'], xerr=results_df['CV Std'], 
         color='#3498db', alpha=0.7, capsize=5)
ax1.set_yticks(x_pos)
ax1.set_yticklabels(results_df['Model'])
ax1.set_xlabel('Accuracy (CV)', fontsize=12)
ax1.set_title('Model Comparison - Cross Validation', fontsize=14, fontweight='bold')
ax1.grid(axis='x', alpha=0.3)
ax1.invert_yaxis()

# Add values on bars
for i, (mean, std) in enumerate(zip(results_df['CV Mean'], results_df['CV Std'])):
    ax1.text(mean + 0.01, i, f'{mean:.3f}Â±{std:.3f}', 
            va='center', fontsize=9, fontweight='bold')

# Plot 2: Train vs Test accuracy
ax2 = axes[1]
x_pos = np.arange(len(results_df))
width = 0.35
ax2.barh(x_pos - width/2, results_df['Train Acc'], width, 
         label='Train', color='#2ecc71', alpha=0.7)
ax2.barh(x_pos + width/2, results_df['Test Acc'], width, 
         label='Test', color='#e74c3c', alpha=0.7)
ax2.set_yticks(x_pos)
ax2.set_yticklabels(results_df['Model'])
ax2.set_xlabel('Accuracy', fontsize=12)
ax2.set_title('Train vs Test Accuracy', fontsize=14, fontweight='bold')
ax2.legend()
ax2.grid(axis='x', alpha=0.3)
ax2.invert_yaxis()

plt.tight_layout()
plt.show()

# Identify overfitting
print("\n" + "="*70)
print("OVERFITTING ANALYSIS")
print("="*70)
results_df['Overfitting'] = results_df['Train Acc'] - results_df['Test Acc']
results_df_sorted = results_df.sort_values('Overfitting', ascending=False)
print("\nModels with the largest Train-Test difference (possible overfitting):")
display(results_df_sorted[['Model', 'Train Acc', 'Test Acc', 'Overfitting']].head(3))

print("\nModels with better generalization:")
display(results_df_sorted[['Model', 'Train Acc', 'Test Acc', 'Overfitting']].tail(3))
print("="*70)
```
